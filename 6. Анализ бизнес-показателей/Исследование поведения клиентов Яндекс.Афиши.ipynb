#### Retention Rate

```python
# Вычислим дату первого визита
first_session_date = visits.groupby(['uid'])['start_ts'].min()
first_session_date.name = 'first_session_date'
```

```python
# Добавим в основную таблицу данные о первом визите
visits = pd.merge(visits, first_session_date, on='uid')
visits['first_session_month'] = visits['first_session_date'].astype('datetime64[M]')
visits['session_month'] = visits['start_ts'].astype('datetime64[M]')
```

```python
# Рассчитаем месяц жизни когорты
visits['cohort_lifetime'] = visits['session_month'] - visits['first_session_month']
visits['cohort_lifetime'] = (visits['cohort_lifetime'] / np.timedelta64(1, 'M')).round().astype(int)
visits['first_session_month'] = visits['first_session_month'].dt.strftime('%Y-%m')

# Посчитаем количество пользователей в когортах
cohorts = visits.groupby(['first_session_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_session_month', 'uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})

# Рассчитаем retention
cohorts = cohorts.merge(initial_users_count, on='first_session_month')
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users']
cohorts = cohorts.query('cohort_lifetime > 0')

# Строим сводную таблицу
retention_pivot = cohorts.pivot_table(
    index='first_session_month', columns='cohort_lifetime', values='retention')

# Построим тепловую карту
print('Cредний Retention Rate на второй месяц «жизни» когорт: {:.1%}'.format(retention_pivot[1].mean()))
sns.set(style='white')
plt.figure(figsize=(13, 9))
plt.title('Retention rate')
sns.heatmap(retention_pivot, annot=True, fmt='.1%', linewidths=1, linecolor='gray')
plt.xlabel('Месяц существования когорты')
plt.ylabel('Когорта'); 
```

Из полученного графика можно сделать следующие выводы:
- на второй месяц существования практически все когорты показывают свой лучший результат;
- к концу исследуемого периода практически все когорты показывают свой худший результат удержания от 2 до 2,7 процентов;
- лучшие показатели удержания пользователей у когорты июня 2017 года

#### Lifetime Value

```python
# В новую таблицу добавим данные о дате и месце первой покупки, месяце остальных покупок
upd2_orders = orders.merge(first_order_date, on='uid')
upd2_orders['first_order_month'] = upd2_orders['first_order_date'].astype('datetime64[M]')
upd2_orders['order_month'] = upd2_orders['buy_ts'].astype('datetime64[M]')

# Посчитаем количество новых покупателей за каждый месяц
cohorts_sizes = upd2_orders.groupby('order_month').agg({'uid':'nunique'}).reset_index()
cohorts_sizes = upd2_orders.groupby('first_order_month').agg({'uid':'nunique'}).reset_index()
cohorts_sizes.columns = ['first_order_month', 'n_buyers']

# Сгруппируем данные по месяцу первой покупки и месяцу каждого заказа и суммируем выручку
cohorts_ltv = upd2_orders.groupby(['first_order_month', 'order_month']).agg({'revenue': 'sum'}).reset_index()

# Вычислим месяц жизни когорт и ltv
report = pd.merge(cohorts_sizes, cohorts_ltv, on='first_order_month')
report['lifetime'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')
report['lifetime'] = report['lifetime'].round().astype('int')
report['ltv'] = report['revenue'] / report['n_buyers']
report['first_order_month'] = report['first_order_month'].dt.strftime('%Y-%m')

# Сделаем срез за полгода и построим сводную таблицу
report_pivot = report.query('first_order_month <= "2017-12-01"').pivot_table(
    index='first_order_month', columns='lifetime', values='ltv')
report_pivot = report_pivot.cumsum(axis=1).round(2)

# Построим тепловую карту
sns.set(style='white')
plt.figure(figsize=(10, 7))
plt.title('LTV')
sns.heatmap(report_pivot, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
plt.yticks(rotation='horizontal')
plt.xlabel('Месяц существования когорты')
plt.ylabel('Когорта'); 
```

```python
report_pivot2 = report.query('first_order_month <= "2017-12-01" and lifetime < 6').pivot_table(
    index='lifetime', columns='first_order_month', values='ltv')
report_pivot2 = report_pivot2.cumsum(axis=0).round(2)
report_pivot2.plot(figsize=(16, 9), grid=True, title='LTV')
plt.xlabel('Месяц существования когорты')
plt.ylabel('Значение ltv');
```

Из полученных данных можно сделать выводы:
- сентябрьская когорта, является самой прибыльной для бизнеса, на 3 месяце существования совершает заметный рывок в сторону увеличения значения ltv и к концу рассматриваемого периода составляет - 12.31; 
- когорта покупателей, совершивших первую покупку в июне и принеся бизнесу самый низкий доход, показала одну из лучших тенденций к росту и концу рассматриваемого периода занимает второе место по прибыльности;
- июльская когорта показывает отнюдь не самые низкие и в тоже время стабильные результаты;
- у октябрьской когорты самые низкие результаты и практически отсутствуют тенденции к росту.

#### Стоимость привлечения одного клиента

```python
# Вычислим дату первой покупки и источника первого визита
buyers = orders.groupby('uid').agg({'dt': 'first'}).rename(columns = {'dt': 'first_order_dt'}).reset_index()
buyers = buyers.merge(visits.sort_values(by = 'start_ts').groupby('uid').agg(
        {'source_id': 'first'}).reset_index(), on = 'uid', how = 'left')

# Группируя по источнику и дате первой покупки установим количество новых покупателей
buyers_daily = buyers.groupby(['source_id', 'first_order_dt']).agg({'uid': 'count'}).reset_index()
buyers_daily.rename(columns={'uid': 'n_buyers'}, inplace = True)

# Объеденив полученые данные с расходами на рекламу, вычислим стоимость привлечения одного клиента
costs_ = pd.merge(buyers_daily, costs, left_on=['source_id', 'first_order_dt'], right_on=['source_id', 'dt'])
costs_['costs_per_buyer'] = costs_['costs']/costs_['n_buyers']
display(costs_.sort_values(by = ['cost_month']).head(5))
```

#### Рентабельность инвестиций в маркетинг

```python
# Определяем источник получения пользователей
users = visits.sort_values(by=['uid', 'start_ts']).groupby('uid').agg({'source_id':'first'}).reset_index()

# Создадим функцию для получения средних значений ROMI и тепловых карт
def romi_from_source(source):
    # Создаем срезы в таблицах по источникам
    source_users = users.query('source_id == @source')['uid']
    orders_ltv = orders.query('uid in @source_users')
    costs_ltv = costs.query('source_id == @source')
    
    # По аналогии с предыдущими заданиями производим необходимые вычисления
    orders_ltv['order_month'] = orders_ltv['buy_ts'].astype('datetime64[M]')
    costs_ltv['month'] = costs_ltv['dt'].astype('datetime64[M]')
    first_orders = orders_ltv.groupby('uid').agg({'order_month': 'min'}).reset_index()
    first_orders.columns = ['uid', 'first_order_month']
    cohort_sizes = (first_orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index())
    cohort_sizes.columns = ['first_order_month', 'n_buyers']
    orders_new = pd.merge(orders_ltv, first_orders, on='uid')
    cohorts = (orders_new.groupby(['first_order_month', 'order_month']).agg({'revenue': 'sum'}).reset_index())
    
    report = pd.merge(cohort_sizes, cohorts, on='first_order_month')
    report = pd.merge(report, costs_ltv.groupby('month')['costs'].sum(), left_on='order_month', right_on='month')

    margin_rate = 1
    report['gp'] = report['revenue'] * margin_rate
    report['age'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M')
    report['age'] = report['age'].round().astype('int')
    report['ltv'] = report['gp'] / report['n_buyers']
    report['cac'] = report['costs'] / report['n_buyers']
    report['romi'] = report['ltv'] / report['cac']

    # Оставим когорты, которые просуществовали не менее полугода
    report2 = report.query('first_order_month <= "2017-12"')
    report2['first_order_month'] = report2['first_order_month'].dt.strftime('%Y-%m')

    # Построим сводные таблицы
    output_romi = report2.pivot_table(
        index='first_order_month', columns='age', values='romi', aggfunc='sum').round(2)
    output_romi = output_romi.cumsum(axis=1).round(2).fillna(0)
    mean_romi = output_romi.mean(axis=0).round(2)
    output_ltv = report2.pivot_table(
        index='first_order_month', columns='age', values='ltv', aggfunc='sum').round(2)
    output_ltv = output_ltv.cumsum(axis=1).round(2).fillna(0)
    mean_ltv = output_ltv.mean(axis=0).round(2)
    
    # Для сравнения рентабельности, возьмем среднее 6го месяца жизни когорт, построим тепловую карту
    print('Среднее значение LTV для пользователей из {}го источника на 6й месяц жизни когорты:'. format(source))
    display(mean_ltv[5])
    sns.set(style='white')
    plt.figure(figsize=(12, 7))
    plt.title('LTV пользователей из {}го источника'.format(source))
    sns.heatmap(output_ltv, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
    plt.yticks(rotation='horizontal')
    plt.xlabel('Месяц существования когорты')
    plt.ylabel('Когорта');
    print('Среднее значение ROMI для пользователей из {}го источника на 6й месяц жизни когорты:'. format(source))
    display(mean_romi[5])
    sns.set(style='white')
    plt.figure(figsize=(12, 7))
    plt.title('ROMI пользователей из {}го источника'.format(source))
    sns.heatmap(output_romi, annot=True, fmt='.2f', linewidths=1, linecolor='gray')
    plt.yticks(rotation='horizontal')
    plt.xlabel('Месяц существования когорты')
    plt.ylabel('Когорта'); 
```

Произведя когортный анализ групп пользователей, пришедших из разных источников рекламы и просуществовавших не менее полугода, для сравнения возьмем среднее значение LTV / ROMI на шестой месяц существования таких групп.

- 1 источник - 12.01 / 1.55
- 2 источник - 15.49 / 1.24
- 3 источник - 5.73 / 0.4
- 4 источник - 6.03 / 0.86
- 5 источник - 9.35 / 1.17
- 9 источник - 5.82 / 1.15
- 10 источник - 1.68 / 0.33

Таким образом, можно сказать, что пользователи, пришедшие из 1, 2, 5, 9 источников к шестому месяцу существования полностью окупились, чего нельзя сказать о когортах, пришедших из 3, 4, 10 источников, привлечение пользователей из которых нерентабельно для бизнеса.
